```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# random marks
np.random.seed(42)
marks = np.random.randint(0, 101, size=(100, 5))
marks
```

```python
#average marks
avg_marks = marks.mean(axis=0)
subjects = ['Math', 'Science', 'English', 'History', 'Art']

marks
```

```python
#Bar chart
plt.figure(figsize=(8, 5))
plt.bar(subjects, avg_marks, color='skyblue')
plt.title('Average Marks per Subject')
plt.ylabel('Average Marks')
plt.xlabel('Subjects')
plt.ylim(0, 100)
plt.show()
```

```python
#histogram
total_marks = marks.sum(axis=1)

plt.figure(figsize=(8, 5))
plt.hist(total_marks, bins=10, color='orange', edgecolor='black')
plt.title('Total Marks Distribution')
plt.xlabel('Total Marks')
plt.ylabel('Number of Students')
plt.show()
```

```python
#piechart
passed = np.all(marks >= 40, axis=1)
num_passed = np.sum(passed)
num_failed = len(passed) - num_passed

plt.figure(figsize=(6, 6))
plt.pie([num_passed, num_failed], labels=['Passed', 'Failed'], autopct='%1.1f%%',
        colors=['green', 'red'], startangle=90)
plt.title('Pass vs Fail Percentage')
plt.show()
```
