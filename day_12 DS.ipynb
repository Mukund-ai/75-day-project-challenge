```python
#Simulating Weather Data Using NumPy
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Mean temperature = 25°C, Standard Deviation = 5°C
mean_temp = 25
std_dev = 5
days = 365

# Generate temperature data
temperature_data = np.random.normal(loc=mean_temp, scale=std_dev, size=days)

# Display first 10 days for verification
print("First 10 days temperatures:", temperature_data[:10])
```

```python
#average temperature
avg_temp = np.mean(temperature_data)
print(f"Average temperature of the year: {avg_temp:.2f}°C")
```

```python
max_temp = np.max(temperature_data)
min_temp = np.min(temperature_data)

print(f"Hottest temperature: {max_temp:.2f}°C")
print(f"Coldest temperature: {min_temp:.2f}°C")
```

```python
#Line Chart of Temperature Over the Year
plt.figure(figsize=(14, 5))
plt.plot(temperature_data, color='orange')
plt.title("Simulated Daily Temperatures Over a Year")
plt.xlabel("Day of the Year")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
#Histogram of Temperatures
plt.figure(figsize=(8, 5))
plt.hist(temperature_data, bins=30, color='skyblue', edgecolor='black')
plt.title("Temperature Distribution")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```
