```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#for marks
np.random.seed(42)
marks=np.random.randint(0,101,size=(100,5))
marks
```

```python
#Dataframe create
df = pd.DataFrame(marks, columns=['Math', 'Science', 'English', 'History', 'Computer'])
df['Student_ID'] = np.arange(1, 101)
df
```

```python
#Reorder
df = df[['Student_ID', 'Math', 'Science', 'English', 'History', 'Computer']]
df.head()
```

```python
#for last lower lines
df = df[['Student_ID', 'Math', 'Science', 'English', 'History', 'Computer']]
df.tail()
```

```python
#for find some operations in it
df = df[['Student_ID', 'Math', 'Science', 'English', 'History', 'Computer']]
df.describe()
```

```python
#analysis it
df['Total'] = df[['Math', 'Science', 'English', 'History', 'Computer']].sum(axis=1)
df['Average'] = df['Total'] / 5

print("\nTop 5 Students by Total Marks:")
print(df.sort_values(by='Total', ascending=False).head(5))
```

```python
#aveage marks per subject
subject_means = df[['Math', 'Science', 'English', 'History', 'Computer']].mean()
subject_means.plot(kind='bar', color='skyblue')
plt.title('Average Marks per Subject')
plt.ylabel('Marks')
plt.show()
```
