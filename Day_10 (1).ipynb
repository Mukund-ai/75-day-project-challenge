```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Fake data
df = pd.DataFrame({
    'Date': pd.date_range('2023-01-01', periods=100, freq='D'),
    'Product': np.random.choice(['A', 'B', 'C'], 100),
    'Revenue': np.random.randint(100, 1000, 100)
})

df['Month'] = df['Date'].dt.month

# Monthly sales
monthly = df.groupby('Month')['Revenue'].sum()
print(monthly)

monthly.plot(kind='line')
plt.title('Monthly Revenue')
plt.show()
```

```python
# 1️⃣ Monthly revenue per product
monthly_product = df.groupby(['Month', 'Product'])['Revenue'].sum().unstack()
print(monthly_product)

monthly_product.plot(kind='bar', stacked=True)
plt.title('Monthly Revenue by Product')
plt.ylabel('Revenue')
plt.xlabel('Month')
plt.show()
```
