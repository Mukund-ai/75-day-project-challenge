```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

num_trials = 10000
```

```python
# Simulate 2 coin tosses (0 = Tail, 1 = Head)
#coin toss
coin_tosses = np.random.choice([0, 1], size=(num_trials, 2))

# Condition: At least one head
at_least_one_head = np.any(coin_tosses == 1, axis=1)

# Empirical Probability
empirical_prob_heads = np.mean(at_least_one_head)

# Theoretical Probability: 1 - P(both tails) = 1 - (0.5 * 0.5) = 0.75
theoretical_prob_heads = 0.75

print(f"Empirical P(at least one head): {empirical_prob_heads:.4f}")
print(f"Theoretical P(at least one head): {theoretical_prob_heads:.4f}")
```

```python
#Dice roll
# Simulate 2 dice rolls (values from 1 to 6)
dice_rolls = np.random.randint(1, 7, size=(num_trials, 2))

# Sum of dice
sums = np.sum(dice_rolls, axis=1)

# Condition: Sum > 9
sum_greater_than_9 = sums > 9

# Empirical Probability
empirical_prob_sum = np.mean(sum_greater_than_9)

# Theoretical Probability:
# Possible sums > 9: (4,6),(5,5),(5,6),(6,4),(6,5),(6,6) = 6 combinations
# Total = 36 combinations â†’ P = 6/36 = 0.1667
theoretical_prob_sum = 6/36

print(f"Empirical P(sum > 9): {empirical_prob_sum:.4f}")
print(f"Theoretical P(sum > 9): {theoretical_prob_sum:.4f}")
```

```python
#Visualisation
# Coin Toss Outcome Distribution
plt.figure(figsize=(6, 4))
labels = ['At Least One Head', 'No Head']
counts = [np.sum(at_least_one_head), num_trials - np.sum(at_least_one_head)]
plt.bar(labels, counts, color=['green', 'red'])
plt.title("Coin Toss Outcome (2 Tosses x 10,000)")
plt.ylabel("Frequency")
plt.show()

# Dice Sum Distribution
plt.figure(figsize=(8, 4))
plt.hist(sums, bins=np.arange(2, 14)-0.5, edgecolor='black')
plt.title("Distribution of Dice Sums (10,000 Trials)")
plt.xlabel("Sum")
plt.ylabel("Frequency")
plt.xticks(np.arange(2, 13))
plt.show()
```
