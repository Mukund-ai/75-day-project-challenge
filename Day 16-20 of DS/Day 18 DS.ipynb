```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, norm, uniform
```

```python
#Binomial distribution
# Parameters
n, p = 10, 0.5
x = np.arange(0, n+1)
binom_probs = binom.pmf(x, n, p)

# Plot
plt.figure(figsize=(7,4))
sns.barplot(x=x, y=binom_probs, color='skyblue')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
# Parameters Noral distributions
mu, sigma = 50, 10
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Plot
plt.figure(figsize=(7,4))
plt.plot(x, y, label="Normal PDF", color='green')
plt.fill_between(x, y, alpha=0.3, color='green')
plt.title("Normal Distribution (mean=50, std=10)")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Simulate "height" data: normal vs uniform
normal_heights = np.random.normal(loc=170, scale=10, size=1000)
uniform_heights = np.random.uniform(low=150, high=190, size=1000)

# Plot histograms
plt.figure(figsize=(10,5))

# Normal
sns.kdeplot(normal_heights, label='Normal Distribution', fill=True, color='blue')

# Uniform
sns.kdeplot(uniform_heights, label='Uniform Distribution', fill=True, color='orange')

plt.title("Normal vs Uniform Distribution (Simulated Heights)")
plt.xlabel("Height (cm)")
plt.legend()
plt.grid(True)
plt.show()
```
