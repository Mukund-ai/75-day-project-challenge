```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)
num_students = 100

# Generate synthetic data
math = np.random.normal(75, 10, num_students)
science = math + np.random.normal(0, 5, num_students)  # highly correlated
english = np.random.normal(70, 12, num_students)
history = english + np.random.normal(0, 8, num_students)
sports = np.random.normal(50, 15, num_students)  # uncorrelated

# Create DataFrame
df = pd.DataFrame({
    'Math': math,
    'Science': science,
    'English': english,
    'History': history,
    'Sports': sports
})

df.head()
```

```
### ðŸ§  Matrix Observations:

- **Covariance matrix** gives raw relationships (sensitive to units).
- **Correlation matrix** standardizes the values (range: -1 to 1).
- Look for values near **+1 or -1** for strong linear relationships.
```

```
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title("ðŸ“Œ Correlation Matrix Heatmap")
plt.show()
```

```
## ðŸ” Strongly Correlated Pairs (Examples):
- `Math` â†” `Science` â†’ High positive correlation
- `English` â†” `History` â†’ Likely moderately strong
- `Sports` â†’ Weak or no correlation with academic subjects

---

## âœ… Summary:
- Use `.corr()` for correlation matrix (Pearson by default)
- Use `.cov()` for raw covariance values
- Visualize with `sns.heatmap()` for quick pattern recognition
```
