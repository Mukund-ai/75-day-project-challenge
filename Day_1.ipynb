```python
import numpy as np
import matplotlib.pyplot as plt

# 1️⃣ Create arrays
zeros_array = np.zeros((3, 4))
ones_array = np.ones((2, 5))
rand_int_array = np.random.randint(1, 100, size=(4, 4))
identity_matrix = np.eye(4)

print("Zeros Array:\n", zeros_array)
print("Ones Array:\n", ones_array)
print("Random Integer Array:\n", rand_int_array)
print("Identity Matrix:\n", identity_matrix)
```

```python
# Reshaped Array
reshaped = rand_int_array.reshape(2,8)
print(reshaped)
```

```python
#operations
print(np.sum(rand_int_array))
```

```python
print("Mean:", np.mean(rand_int_array))
print("Std Dev:", np.std(rand_int_array))
print("Transpose:\n", rand_int_array.T)
```

```python
#Mastering an dfiltering
filtered = rand_int_array[rand_int_array > 50]
print("Elements > 50:", filtered)
```

```python
#Graph plot
x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.title("Simple Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
